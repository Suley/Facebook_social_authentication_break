```python
import face_recognition
import cv2
import os
```

```python
face_recognition.compare_faces
# 照片路径
friends_images_path = './train_img/'
predict_images_path  = './predict_images2/'

# 已知照片的人脸编码列表
known_face_encodings = []
#名字列表
known_face_names = []

# 遍历路径下的文件
for filename in os.listdir(friends_images_path):
    # 忽略macos 的 .DS_Store
    if filename.startswith('.'):
        continue
    image_path = friends_images_path + '/' + filename
    print (image_path)
    friend_image = face_recognition.load_image_file(image_path)
    # 只获取第一个人的 人脸编码 和 名称
    #print (face_recognition.face_encodings(friend_image))
    #break
    #print(face_recognition.face_encodings(friend_image))
    
    friend_face_encoding = face_recognition.face_encodings(friend_image)
    if len(friend_face_encoding) != 0:
        friend_face_encoding = friend_face_encoding[0]
        known_face_encodings.append(friend_face_encoding)
        known_face_names.append(filename.split('.')[0])
print (known_face_names)
```

```python
def sort_dict(dict_words):
    """
    字典排序
    :param dict_words:
    :return:
    """
    keys = dict_words.keys()
    values = dict_words.values()
    list_one = [(key, val) for key, val in zip(keys, values)]
    list_sort = sorted(list_one, key=lambda x: x[1], reverse=False)
    return list_sort
```

```python

predict_images_path = './clean_img/' 
for filename in os.listdir(predict_images_path):
     # 忽略.DS_Store
    if filename.startswith('.'):
        continue
    image_path = predict_images_path + '/' + filename
    print (image_path)
    predict_image = face_recognition.load_image_file(image_path)
    predict_face_encodings = face_recognition.face_encodings(predict_image)
   # print(predict_face_encodings)
    #存储每一张脸的位置信息
    face_locations = face_recognition.face_locations(predict_image)
    # 遍历一张图中多个脸 编码信息
    for i in range(len(predict_face_encodings)):
        predict_encoding = predict_face_encodings[i]
        face_location = face_locations[i]
        top, right, bottom,left = face_location
        # 在照片中标注人脸
        cv2.rectangle(predict_image,(left,top),(right,bottom),(0,255,0),2)
        # 将已知人脸的编码，和未知人脸的编码作比较，结果存入results数组中
        # results 数组中的每一个元素都是True 或者 False
        # results 数组与已知人脸一一对应
        face_distances = face_recognition.face_distance(known_face_encodings, predict_encoding)
        ans = {}
        for i, face_distance in enumerate(face_distances):
            #print("The test image has a distance of {:.2} from known image #{}".format(face_distance, i))
            #print("- With a normal cutoff of 0.6, would the test image match the known image? {}".format(face_distance < 0.6))
            #print("- With a very strict cutoff of 0.5, would the test image match the known image? {}".format(face_distance < 0.5))
            name = known_face_names[i]
            ans[name] = face_distance
            #print(name,face_distance)
        ans = sort_dict(ans)
        # 方法1
        print ('遍历列表方法1：')
        for i in ans:
            print ("序号：%s  值：%s" % (ans.index(i) + 1, i[0]))
        
        cnt = 0
        for key in ans:  
            cnt += 1
            if cnt == 1:
                print('识别结果',key)
                print('备选项：')
                continue
            print(key)
            if cnt == 6:
                break
            
#         results = face_recognition.compare_faces(known_face_encodings,predict_encoding)
#         for j in range (len(results)):
#             if results[j]:
#                 name = known_face_names[j]
#                 print ('the picture is '+ filename+ ", predicted the people is "+name)
#                 # 在图片中标注姓名
#                 cv2.putText(predict_image,name,(left-10,top-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
#                 break
#     predict_image_rgb = cv2.cvtColor(predict_image, cv2.COLOR_BGR2RGB)
#     cv2.imshow("Output",predict_image_rgb)
#     cv2.waitKey(0)
```
